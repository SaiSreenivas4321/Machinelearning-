1.0 Getting Independent features in a dataset and Dependent feature in Series object

```python
dataset.head()
```

```python
X= dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

## 2.0 Splitting data into training and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=10)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

3.0 Feature engineering 3.1 Standardisation/ feature scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
### using fit_transform to Standardize the data
X_train=scaler.fit_transform(X_train)
X_train
```

```python
### here using transform only to avoid data leakage
### (training mean and training std will be used for standardisation when we use transform)
X_test=scaler.transform(X_test)
X_test
```

##4.0 Model 1.0 Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
### Passing independant and dependant training data to the model
regression.fit(X_train,y_train)
```

```python
### print the coefficients and the intercept of the best fit hyperplane
print("1. Co-efficients of Independent features {}".format(regression.coef_))
print("2. Intercept of best fit hyper-plane {}".format(regression.intercept_))
```

### 1.1 Using Above Model to get prediction for test data

```python

reg_pred=regression.predict(X_test)
reg_pred
```

### 1.2 Assumptions of linear regre
1.2.1 Linear relationship 
1. Test truth data and Test predicted data sh2. This is an indication of a good model.

```python
plt.scatter(y_test, reg_pred)
plt.xlabel('Test Truth data')
plt.ylabel('Test predicted data')
```

### 1.2.2 Residual distribution 
1. Residuals should follow normal distribution.
2. If residuals follow normal distribution, it indicates we have a good model.

```python
residuals=y_test-reg_pred
residuals.head()
```

```python
sns.displot(residuals, kind='kde')
```

### 1.2.3 Uniform distribution 
1. Residuals vs Predictions should follow a uniform distribution.
2. If Residuals vs Predictions follow uniform distribution, it indicates we have a good model.

```python
plt.scatter(reg_pred, residuals)
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
```

### Perofirmance Metrices 
### 1.3.0 Performance matrix 
1.3.1 Cost Function values [ ]:


```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
#from sklearn.metrics import root_meansquare_error
```

### 1.3.1.1 Mean Squared Error 

```python

print("Mean Squared error is {}".format(round(mean_squared_error(y_test, reg_pred),2)))
```

### 1.3.1.2 Mean Absolute error

```python
print("Mean abosulte error is {}".format(round(mean_absolute_error(y_test,reg_pred),2)))
```

```python

print("Root Mean Squared error is {}".format(round(np.sqrt(mean_squared_error(y_test, reg_pred)),2)))
```

### 2.3.2 R Square and Adjusted R Square 
2.3.2.1 R square Value

```python

from sklearn.metrics import r2_score
```

```python
r_score=r2_score(y_test, reg_pred)
print("Our Model has {} percent accuracy".format(round(r_score*100,2)))
```

### Adjusted R_square 

```python
adj_r_score=1-(1-r_score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print("Adjusted R square accuracy is {} percent".format(round(adj_r_score*100,2)))
```
